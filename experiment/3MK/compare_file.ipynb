```python
import csv
import json


# CSV 파일 경로
csv_file_path_f1 = '/data/ephemeral/home/IR/code/sample_submission.csv'
csv_file_path_f2 = '/data/ephemeral/home/IR/code/sample_submission2.csv'
csv_file_path_f3 = '/data/ephemeral/home/IR/code/sample_submission3.csv'
csv_file_path_f4 = '/data/ephemeral/home/IR/code/sample_submission5.csv'
#csv_file_path_f1 = '/data/ephemeral/home/IR/code/sample_submission2.csv'

# 파일 경로 추가하면, 리스트에 추가. 
csv_file_path_list = [csv_file_path_f1, csv_file_path_f2, csv_file_path_f3, csv_file_path_f4]
# CSV 파일 읽기
######################### 초기 설정 부분###############################################
result_standalone_query= []
result_answer = []
with open(csv_file_path_f1) as f:
    for l_1 in f:
        line_one = json.loads(l_1)
        result_standalone_query.append(line_one["standalone_query"])
        result_answer.append(line_one["answer"])
######################################################################################

total_references = [[]*_list for _list in range(len(csv_file_path_list))]
for idx, path in enumerate(csv_file_path_list):
    _total_references = []
    with open(path) as file_one:
        for line in file_one:
            _total = []
            line_one = json.loads(line)
            try:
                _total.append(line_one["references"][0]["content"])
                _total.append(line_one["references"][1]["content"])
                _total.append(line_one["references"][2]["content"])
            except:
                _total.append(-1)
        
            _total_references.append(_total)
        #print(idx, _total_references)
    total_references[idx].append(_total_references)
```

```python
count = 0
for i in range(len(result_standalone_query)):
    if count == 5:
        break
    display(result_standalone_query[i]) # 쿼리 문 
    display(result_answer[i]) # 답변 
    for num in range(len(csv_file_path_list)): 
        display(total_references[num][0][i])  # num: csv number, i: eval_number(220)
    
    count += 1
```
